```python
def plot_magnitude_spectrum(signal, title, sr):
    FouTr = np.fft.fft(signal)
    magnitude_spectrum = np.abs(FouTr)
    
    #plot magnitude spectrum
    plt.figure(figsize=(18, 5))
    
    frequency = np.linspace(0, sr, len(magnitude_spectrum))
    plt.plot(frequency, magnitude_spectrum)
    plt.xlabel("Frequency (Hz)")
    plt.title(title)
    
    plt.show()
```

```python
plot_magnitude_spectrum(piano_c4, "piano", sr)
```

```python
def plot_magnitude_spectrum(signal, title, sr, f_ratio=1):
    FouTr = np.fft.fft(signal)
    magnitude_spectrum = np.abs(FouTr)
    
    #plot magnitude spectrum
    plt.figure(figsize=(18, 5))
    
    frequency = np.linspace(0, sr, len(magnitude_spectrum))
    num_frequency_bins = int(len(frequency) * f_ratio)
    
    plt.plot(frequency[:num_frequency_bins], magnitude_spectrum[:num_frequency_bins])
    plt.xlabel("Frequency (Hz)")
    plt.title(title)
    
    plt.show()
```

```python
plot_magnitude_spectrum(piano_c4, "piano", sr, 0.5)
```

```python
plot_magnitude_spectrum(piano_c4, "piano", sr, 0.1) # The 1st harmonic has more energy than the fundamental one
```
